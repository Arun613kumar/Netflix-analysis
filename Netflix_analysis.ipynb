# **NETFLIX STOCK ANALYSIS PRODUCT**



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df=pd.read_csv("/Netflix.csv")
df
```

```python
sns.set(rc={'figure.figsize':(10,5)})
```

# *set index number to date*

```python
df['Date']=pd.to_datetime(df['Date'])
df=df.set_index('Date')
df.head()
```

# **1).how much volume of stock is traded **

```python
sns.lineplot(x=df.index, y=df['Volume'],label='Volume')
plt.title('volume of stock versus time')
plt.show()
```

# **2)NETFLIX STOCK PRICE-HIGH,OPEN,CLOSE**









```python
df.plot(y=['Open','Close','High'])
plt.title("Netflix Stock Price")
plt.show
```

# ***3)NETFLIX STOCK PRICE-DAY,MONTH,YEAR WISE***

```python
fig, (ax1,ax2,ax3)=plt.subplots(3,figsize=(10,10))
df.groupby(df.index.day).mean().plot(y='Volume',ax=ax1,xlabel='Day')
df.groupby(df.index.month).mean().plot(y='Volume',ax=ax2,xlabel='month')
df.groupby(df.index.year).mean().plot(y='Volume',ax=ax3,xlabel='year')
plt.show()
```

## 4)DATE WITH HIGH STOCK PRICE

```python
a=df.sort_values('High',ascending=False).head(5)
a['High']
```

# **5)DATE WITH LOW STOCK PRICE**

```python
b=df.sort_values('Low',ascending=True).head(5)
b['Low']
```

```python
fig,axes=plt.subplots(nrows=1, ncols=2, sharex=True,figsize=(12,5))
fig.suptitle('High & Low vlues stock period of time')
sns.lineplot(ax=axes[0],y=df['High'],x=df.index,color='green')
sns.lineplot(ax=axes[1],y=df['Low'],x=df.index,color='red')
```

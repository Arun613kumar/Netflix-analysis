**18.Classify Movies Based on Ratings [Excellent, Good, and Average]**

```python
df.columns
```

```python
def rating (rating):
  if rating>=7.0:
    return 'Excellent'
  elif rating>=6.0:
    return 'Good'
  else:
    return 'Average'
```

```python
df['rating_cat']=df['Rating'].apply(rating)
```
